# Supervised Learning: Regression Models
This notebook implements Linear Regression and Polynomial Regression on a real estate dataset. It evaluates model performance using R² score and RMSE, and visualizes regression results.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)
```

## Load and Prepare Data
Load the real estate dataset, select features, define the target, and split into training and test sets.

```python
# Load dataset
file_path = "real_estate_dataset.csv"
df = pd.read_csv(file_path)

# Display dataset overview
print("Dataset shape:", df.shape)
print(df.head())

# Define features and target
feature_columns = [
    "Square_Feet",
    "Num_Bedrooms",
    "Num_Bathrooms",
    "Num_Floors",
    "Year_Built",
    "Has_Garden",
    "Has_Pool",
    "Garage_Size",
    "Location_Score",
    "Distance_to_Center",
]
X = df[feature_columns]
y = df["Price"]

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
print("Training samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])
```

## Train Linear Regression Model
Fit a Linear Regression model on the training data.

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Make predictions on test set
y_pred_linear = linear_model.predict(X_test)
print("Linear regression model trained.")
```

## Train Polynomial Regression Model
Use PolynomialFeatures to transform the input features and fit a LinearRegression model to the transformed data.

```python
poly_degree = 2
poly_transformer = PolynomialFeatures(degree=poly_degree, include_bias=False)
X_train_poly = poly_transformer.fit_transform(X_train)
X_test_poly = poly_transformer.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

y_pred_poly = poly_model.predict(X_test_poly)
print(f"Polynomial regression model trained with degree {poly_degree}.")
print("Polynomial feature count:", X_train_poly.shape[1])
```

## Evaluate Models with R² and RMSE
Compute R² score and RMSE for both models on the test data.

```python
# Evaluation function

def evaluate_model(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{name} Results")
    print("R2 score:", round(r2, 4))
    print("RMSE:", round(rmse, 2))
    print("-")

# Evaluate both models
evaluate_model("Linear Regression", y_test, y_pred_linear)
evaluate_model("Polynomial Regression", y_test, y_pred_poly)
```

## Visualize Regression Results
Plot actual vs predicted values and compare the regression performance for both models.

```python
# Create a comparison DataFrame for plotting
comparison_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted_Linear": y_pred_linear,
    "Predicted_Polynomial": y_pred_poly,
})
comparison_df = comparison_df.reset_index(drop=True)

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.scatter(comparison_df["Actual"], comparison_df["Predicted_Linear"],
            alpha=0.7, label="Linear Regression", edgecolor="w")
plt.scatter(comparison_df["Actual"], comparison_df["Predicted_Polynomial"],
            alpha=0.7, label="Polynomial Regression", edgecolor="w")
plt.plot([comparison_df["Actual"].min(), comparison_df["Actual"].max()],
         [comparison_df["Actual"].min(), comparison_df["Actual"].max()],
         color="black", linestyle="--", label="Perfect prediction")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.legend()
plt.tight_layout()
plt.show()

# Plot residuals for both models
comparison_df["Residual_Linear"] = comparison_df["Actual"] - comparison_df["Predicted_Linear"]
comparison_df["Residual_Polynomial"] = comparison_df["Actual"] - comparison_df["Predicted_Polynomial"]

plt.figure(figsize=(12, 6))
plt.hist(comparison_df["Residual_Linear"], bins=30, alpha=0.6, label="Linear Residuals")
plt.hist(comparison_df["Residual_Polynomial"], bins=30, alpha=0.6, label="Polynomial Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.title("Residual Distribution for Linear and Polynomial Models")
plt.legend()
plt.tight_layout()
plt.show()
```
